```python
import numpy as np
from gtts import gTTS
import os
```

```python
def load_text(filename):
    with open(filename, encoding='utf-8') as f:
        return f.read().lower()
```

```python
data = load_text("data.txt")
print(data)
```

```python
def generateTransitionTable(data, k=4):
    T = {}
    for i in range(len(data)-k):
        X = data[i:i+k]
        Y = data[i+k]
        
        if T.get(X) is None:
            T[X] = {}
            T[X][Y] = 1
        elif T[X].get(Y) is None:
            T[X][Y] = 1
        else:
            T[X][Y] += 1
    
    return T
```

```python
# T = generateTransitionTable("Hello Hello Helli")
T = generateTransitionTable(data)
```

```python
def convertTransitionToProbabilities(T):
    
    for kx in T.keys():
        norm_factor = sum(T[kx].values())
        for val in T[kx]:
            T[kx][val] = T[kx][val]/norm_factor
    return T
```

```python
T = convertTransitionToProbabilities(T)
print(T)
```

```python
def predict(ctx, T):
    if T.get(ctx) is None:
        return " "
    possible_keys = list(T[ctx].keys())
    possible_values = list(T[ctx].values())
#     print(possible_keys)
#     print(possible_values)
    return np.random.choice(possible_keys, p=possible_values)
```

```python
predict("Hell", T)
```

```python
k = 4
starting_sent = " कौन"

for i in range(100):
    next_char = predict(starting_sent[-k:], T)
    starting_sent += next_char
    

print(starting_sent)
```

```python
myobj = gTTS(text=starting_sent, lang='hi')
myobj.save("test.mp3")
```
