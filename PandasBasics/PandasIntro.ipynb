```python
titles = ds.get('movie_title')
print(type(titles))
```

```python
freq_title = {}
for title in titles:
    leng = len(title)
    try:
        freq_title[leng] += 1
    except:
        freq_title[leng] = 1
print(freq_title)
```

```python
l = list(freq_title.keys())
a = np.asarray(l)
print(a.min(), a.max())
```

```python
plt.scatter(freq_title.keys(), freq_title.values())
plt.show()
```
